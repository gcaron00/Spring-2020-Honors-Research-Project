## Suppor Vector Machines

Regression gives numerical value classification classifies to group

SVC and NuSVC are similar methods, but accept slightly different sets of parameters and have different mathematical formulations (see section Mathematical formulation). On the other hand, LinearSVC is another implementation of Support Vector Classification for the case of a linear kernel. Note that LinearSVC does not accept keyword kernel, as this is assumed to be linear. It also lacks some of the members of SVC and NuSVC, like support_.

As other classifiers, SVC, NuSVC and LinearSVC take as input two arrays: an array X of size [n_samples, n_features] holding the training samples, and an array y of class labels (strings or integers), size [n_samples]



n problems where it is desired to give more importance to certain classes or certain individual samples keywords class_weight and sample_weight can be used.

**Unbalanced**

SVC (but not NuSVC) implement a keyword class_weight in the fit method. Its a dictionary of the form {class_label : value}, where value is a floating point number > 0 that sets the parameter C of class class_label to C * value.


SVC, NuSVC, SVR, NuSVR, LinearSVC, LinearSVR and OneClassSVM implement also weights for individual samples in method fit through keyword sample_weight. Similar to class_weight, these set the parameter C for the i-th example to C * sample_weight[i].



**Kernel**: Functions used to explicity transform data in raw representation into feature vector representations

**Support vectors**: Data points that are closer to the hyperplane and influence the position and orientation of the hyperplane



**C value:** is determines how far out the classifier searches for support vectors. This is demonstrated by the dashed lines in the above image

* SVR (regression) uses all data points for support vectors


## Support Vectors Classification Basic Single Class 

```python
from sklearn import svm
X = [[0, 0], [1, 1]]
y = [0, 1]
clf = svm.SVC(gamma='auto')
clf.fit(X, y)
```

Above code sets up svm

```python
clf.predict([[2., 2.]])
```

Above code used to predict set of new values

```python
# get support vectors
clf.support_vectors_
```

```python
# get indices of support vectors
clf.support_
```

```python
# get number of support vectors for each class
clf.n_support_
```

## Support Vectors Classification Basic Multi Class

```python
X = [[0], [1], [2], [3]]
Y = [0, 1, 2, 3]
clf = svm.SVC(decision_function_shape='ovo', gamma = 'auto')
clf.fit(X, Y)
```

```python
dec = clf.decision_function([[1]])
dec.shape[1] # 4 classes: 4*3/2 = 6
```

```python
clf.decision_function_shape = "ovr"
dec = clf.decision_function([[1]])
dec.shape[1] # 4 classes
```

If n_class is the number of classes, then n_class * (n_class - 1) / 2 classifiers are constructed and each one trains data from two classes. To provide a consistent interface with other classifiers, the decision_function_shape option allows to monotically transform the results of the one-against-one classifiers to a decision function of shape (n_samples, n_classes).

```python
lin_clf = svm.LinearSVC()
lin_clf.fit(X, Y)
```

```python
dec = lin_clf.decision_function([[1]])
dec.shape[1]
4
```

LinearSVC implements one-vs-the-rest multi-class strategy, thus training n_class models. If there are only two classes, only one model is trained

## Support Vector Regression 

```python
from sklearn import svm
X = [[0, 0], [2, 2]]
y = [0.5, 2.5]
clf = svm.SVR()
clf.fit(X, y)
```

```python
clf.predict([[1, 1]])
```

The method of Support Vector Classification can be extended to solve regression problems. This method is called Support Vector Regression.

The model produced by support vector classification (as described above) depends only on a subset of the training data, because the cost function for building the model does not care about training points that lie beyond the margin. Analogously, the model produced by Support Vector Regression depends only on a subset of the training data, because the cost function for building the model ignores any training data close to the model prediction.

## Support Vector Classification Extended w/ Pandas

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
df0.head()
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'green', marker = '+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'blue', marker = '+')
```
